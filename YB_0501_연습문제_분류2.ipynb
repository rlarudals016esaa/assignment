3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
```

```python
# 1. 딕셔너리를 활용하여 직접 매핑 (white: 0, red: 1)
train['type'] = train['type'].map({'white': 0, 'red': 1})

# 2. 변환 결과 확인
print(train[['type']].head())
print("\n--- 데이터 타입 확인 ---")
print(train['type'].value_counts())
```

```python
# 1. 딕셔너리를 활용하여 직접 매핑 (white: 0, red: 1)
test['type'] = test['type'].map({'white': 0, 'red': 1})

# 2. 변환 결과 확인
print(test[['type']].head())
print("\n--- 데이터 타입 확인 ---")
print(test['type'].value_counts())
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.

# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
train_x = train.drop(columns=['index', 'quality'])

# train_y : train의 'quality'를 저장합니다.
train_y = train['quality']

# test_x : test의 'index'를 제거하고 저장합니다.
test_x = test.drop(columns=['index'])
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
from lightgbm import LGBMClassifier
lgbm_model = LGBMClassifier()
lgbm_model.fit(train_x, train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
pred = lgbm_model.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
smpl_sub['quality'] = pred
```

# 2. XGBoost

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
# 'type' 컬럼을 원-핫 인코딩하고, 결과 컬럼을 정수형(0 또는 1)으로 변환합니다.
train = pd.get_dummies(train, columns=['type'], dtype=int)
test = pd.get_dummies(test, columns=['type'], dtype=int)
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델 정의
model = XGBClassifier()
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X = train.drop('quality', axis=1)
y = train['quality']

# XGBoost는 타겟 변수가 0부터 시작하는 정수형이어야 하므로, quality 값을 0-인덱싱으로 변환합니다.
y = y - y.min()

# fit 메소드를 이용해 모델 학습
model.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측
pred = model.predict(test)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
sample_submission = pd.read_csv('data/sample_submission.csv')
sample_submission['quality'] = pred
sample_submission
```

```python
# 정답파일 내보내기
sample_submission.to_csv('xgb_pred.csv',index = False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
display(train.head())
display(test.head())
```

```python
# 원핫 인코딩 (pd.get_dummies())
train = pd.get_dummies(train, columns=['type'], dtype=int)
test = pd.get_dummies(test, columns=['type'], dtype=int)

# columns=['type']: 원-핫 인코딩을 적용할 컬럼을 지정
#dtype=int: 새로 생성될 더미(dummy) 변수 컬럼의 데이터 타입을 정수형(0 또는 1)으로 지정
```

```python
display(train.head())
display(test.head())
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
lgbm_model = LGBMClassifier()
xgb_model = XGBClassifier()
rf_model = RandomForestClassifier()
# VotingClassifier 정의
voting_model = VotingClassifier(estimators=[('lgbm', lgbm_model), ('xgb', xgb_model), ('rf', rf_model)], voting='soft')
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
X = train.drop('quality', axis=1)
# y 는 train의 qulity 변수
y = train['quality']

# fit 메소드를 이용해 모델 학습
voting_model.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
pred = voting_model.predict(test)
```

```python
sample_submission_voting = pd.read_csv('data/sample_submission.csv')
sample_submission_voting['quality'] = pred
sample_submission_voting
```
